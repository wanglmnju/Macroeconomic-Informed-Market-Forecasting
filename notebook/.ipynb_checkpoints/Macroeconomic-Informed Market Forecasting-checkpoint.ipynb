## Forecasting NASDAQ 5-Day Forward Returns Using Macroeconomic Indicators

### Problem definition

<h2 style="color:blue">Objective:</h2>

Predict 5-day forward log returns of the NASDAQ Composite Index using only economic time-series data that was available in real-time, with a focus on Federal Reserve indicators, Treasury yields, employment data, inflation, and market volatility.

**Mathematically:**

For each trading day $t$, predict:
$$\text{Target}_t = \log(P_{t+6}) - \log(P_{t+1})$$

(6-day ahead price vs. 1-day ahead price, giving a 5-day forward return with 1-day lag)

**<h2 style="color:blue">Constraints:</h2>**
- No future data leakage
- Must handle revised data using ALFRED
- Treat zero values in NASDAQCOM as NULLs and drop them
- Minimum Features: ["FEDFUNDS", "DGS2", "DGS10", "PAYEMS", "CPIAUCSL", "VIXCLS"]

**<h2 style="color:blue">Why This Matters:**

- Trading Strategy Development

Predict short-term market movements to inform tactical asset allocation or hedging decisions.

- Macro-Risk Assessment

Quantify how Federal Reserve policy (rates), Treasury markets (yield curve), and economic health (NFP/CPI) impact tech stock performance.

- Data Challenges

Real-world constraints: Avoid look-ahead bias by using only data available at prediction time (via ALFRED's vintage data).

**<h2 style="color:blue">Business Impact:**
    
A successful model could:

- For Asset Managers: Signal short-term rebalancing opportunities in tech-heavy portfolios.

- For Risk Teams: Provide early warnings when macro conditions (e.g., yield curve inversion) predict NASDAQ drawdowns.

- For DAI: Demonstrate how real-time economic data pipelines can generate alpha in systematic strategies.

## 1 Data Collection

### feature description:

- NASDAQCOM: The daily closing price of the NASDAQ Composite Index, representing a broad benchmark of U.S. tech and growth stocks. -> daily

- FEDFUNDS: The Federal Funds Rate, set by the Federal Reserve, which influences short-term interest rates and overall monetary policy. -> daily, monthly release. 

- DGS2: The 2-Year U.S. Treasury Yield, a measure of short-term government borrowing costs and investor expectations for interest rates. -> daily

- DGS10: The 10-Year U.S. Treasury Yield, often used as a benchmark for long-term interest rates and economic outlook. -> daily

- PAYEMS: Nonfarm Payrolls, a monthly estimate of the number of jobs added or lost in the U.S. economy, excluding farm workers — a key labor market indicator. -> monthly

- CPIAUCSL: The Consumer Price Index for All Urban Consumers, which measures inflation by tracking changes in the price of a basket of goods and services. -> monthly

- VIXCLS: The CBOE Volatility Index (VIX), often called the “fear index,” which reflects market expectations for near-term volatility in the S&P 500. -> daily


### Fetching data from FRED

```python
# Initialize FRED API (get free key: https://fred.stlouisfed.org/docs/api/api_key.html)
fred = Fred(api_key='FRED key')
```

```python
# Define date range (adjust based on data availability)
start_date = '2000-01-01'  # VIX start date: 1990-01-02
end_date = '2024-12-31'  # Ensure Data Revisions Are Stable
```

```python
# Fetch daily data from FRED
def fetch_fred_data(series_id):
    return fred.get_series(series_id, start_date, end_date)
```

```python
data = pd.DataFrame({
    'NASDAQ': fetch_fred_data('NASDAQCOM'),
    'FEDFUNDS': fetch_fred_data('FEDFUNDS'),
    'DGS2': fetch_fred_data('DGS2'),
    'DGS10': fetch_fred_data('DGS10'),
    'VIX': fetch_fred_data('VIXCLS')})
```

```python
data.info()
```

### Forward filling FEDFUNDS,DGS2,DGS10, VIX

```python
# Step 1: Identify release dates (first biz day of month)
release_dates = (
    pd.date_range(start_date, end_date, freq='MS')  # Business Month Start
    .intersection(data.FEDFUNDS.index)  # Align with actual data points
)
```

```python
# Step 2: Create release date -> value mapping
release_map = pd.Series(data.FEDFUNDS.loc[release_dates].values, index=release_dates)
```

```python
# Step 3: Forward-fill to all days until next release
fedfunds_ptit = fedfunds_ptit = (
    release_map.reindex(data.FEDFUNDS.index, method='ffill')
    .rename('FEDFUNDS_known1')
)
```

```python
# Step 4: Merge with NASDAQ and add 1-day lag
data = data.join(fedfunds_ptit, how='left')
```

### lag for point-in time logic

```python
data.loc[:, 'FEDFUNDS_known'] = data['FEDFUNDS_known1'].shift(1)
data
```

```python
data['DGS2_known'] = data['DGS2'].ffill(limit=2).shift(1)
```

```python
data['DGS10_known'] = data['DGS10'].ffill(limit=2).shift(1)
```

```python
data['VIX_known'] = data['VIX'].ffill(limit=2).shift(1)
```

```python
data
```

### Fetch ALFRED vintages for PAYEMS and CPI

```python
## Transformation to Standard Format
# Load ALFRED file
df = pd.read_csv("PAYEMS_vintages_starting_2000-01-07.csv", parse_dates=['observation_date'])

# Melt to long format
payems_long = df.melt(
    id_vars='observation_date',
    var_name='vintage_date_str',
    value_name='PAYEMS'
).dropna()

# Extract dates from column names (e.g., "PAYEMS_20000107" → 2000-01-07)
payems_long['vintage_date'] = pd.to_datetime(
    payems_long['vintage_date_str'].str.replace('PAYEMS_', ''),
    format='%Y%m%d'
)

# Keep only needed columns
payems_clean = payems_long[['observation_date', 'vintage_date', 'PAYEMS']]
```

```python
payems_df = payems_long.sort_values(['observation_date', 'vintage_date'])[
    ['observation_date', 'vintage_date', 'PAYEMS']]
payems_df = payems_df.reset_index(drop=True)
payems_df
```

```python
## Transformation to Standard Format
# Load your ALFRED file
df = pd.read_csv("CPI_vintages_starting_2000-01-14.csv", parse_dates=['observation_date'])

# Melt to long format
CPI_long = df.melt(
    id_vars='observation_date',
    var_name='vintage_date_str',
    value_name='CPI'
).dropna()

# Extract dates from column names (e.g., "PAYEMS_20000107" → 2000-01-07)
CPI_long['vintage_date'] = pd.to_datetime(
    CPI_long['vintage_date_str'].str.replace('CPIAUCSL_CH1_', ''),
    format='%Y%m%d'
)

# Keep only needed columns
CPI_clean = CPI_long[['observation_date', 'vintage_date', 'CPI']]
```

```python
CPI_df = CPI_long.sort_values(['observation_date', 'vintage_date'])[
    ['observation_date', 'vintage_date', 'CPI']]
CPI_df = CPI_df.reset_index(drop=True)
CPI_df
```

```python
data_filter=data[~data['NASDAQ'].isnull() & data['NASDAQ']>0]
data_filter.info()
```

### merge with Point-in -Time Logic

observation_date should match the month prior to prediction_date because:
- June payrolls affect July predictions
- July payrolls affect August predictions

```python
# Set observation_date to PREVIOUS month
data_filter['observation_date'] = (data_filter.index - pd.DateOffset(months=1)).to_period('M').to_timestamp()
data_filter.head()
```

```python
data_with_payems = pd.merge_asof(
        left=data_filter.sort_index().reset_index()
    .rename(columns={'index': 'prediction_date'}),
        right=payems_df.sort_values('vintage_date'),
        left_on='prediction_date', # Uses the prediction dates as the merge key in the left DataFrame
        right_on='vintage_date', Matches against PAYEMS release dates in the right DataFrame
        by='observation_date',  # group by 
        direction='backward',   # take the last release where vintage_date <= prediction_date
        tolerance=pd.Timedelta(days=90)
    )
```

```python
data_with_payems=data_with_payems[['NASDAQ', 'prediction_date', 'FEDFUNDS_known', 'VIX_known', 'DGS2_known','DGS10_known','observation_date','PAYEMS' ]]
```

```python
merged = pd.merge_asof(
        left=data_with_payems.sort_values('prediction_date'),
        right=CPI_df.sort_values('vintage_date'),
        left_on='prediction_date',
        right_on='vintage_date',
        by='observation_date',
        direction='backward',
        tolerance=pd.Timedelta(days=90)
    )
```

```python
merged.head(10)
```

```python
merged['CPI'] = merged.groupby('observation_date')['CPI'].ffill().shift(1)
merged['PAYEMS'] = merged.groupby('observation_date')['PAYEMS'].ffill().shift(1)
```

## 3. Feature engineering
Transform raw rates into economically meaningful spreads/orthogonal features:

- Term Spread: Captures market expectations (inverted curve = recession signal)

- Policy Spread: Measures how much markets "believe" the Fed (e.g., negative spread = tight policy)

```python
# Yield curve features
final_data['term_spread'] = final_data['DGS10_known'] - final_data['DGS2_known']  # Steepness predictor of recessions
final_data['policy_spread'] = final_data['DGS2_known'] - final_data['FEDFUNDS_known']  # Front-end policy sensitivity
final_data.head(5)
```

```python
# Drop raw rates or keep only FEDFUNDS
data_final=final_data.copy()
data_final.drop(['DGS2_known', 'DGS10_known','PAYEMS','CPI'], axis=1, inplace=True)
data_final.head()
```

## Feature-Specific Scaling

### PAYEMS

```python
# Compute expanding mean/std up to t-1 for each row
X_train['PAYEMS_mean'] = X_train['PAYEMS'].expanding().mean().shift(1)
X_train['PAYEMS_std'] = X_train['PAYEMS'].expanding().std().shift(1)

# Scale PAYEMS using time-aware statistics
X_train['PAYEMS_scaled'] = (X_train['PAYEMS'] - X_train['PAYEMS_mean']) /X_train['PAYEMS_std']
```

```python
# Use training set's final mean/std for static models
X_test_payems_mean = X_train['PAYEMS_mean'].iloc[-1]
X_test_payems_std = X_train['PAYEMS_std'].iloc[-1]
X_test['PAYEMS_scaled'] = (X_test['PAYEMS'] - X_test_payems_mean) / X_test_payems_std
```

### CPI

```python
# Compute expanding mean/std up to t-1 for each row Of CPI
X_train['CPI_mean'] = X_train['CPI'].expanding().mean().shift(1)
X_train['CPI_std'] = X_train['CPI'].expanding().std().shift(1)

# Scale PAYEMS using time-aware statistics
X_train['CPI_scaled'] = (X_train['CPI'] - X_train['CPI_mean']) /X_train['CPI_std']
```

```python
# Use training set's final mean/std for static models
X_test_cpi_mean = X_train['CPI_mean'].iloc[-1]
X_test_cpi_std = X_train['CPI_std'].iloc[-1]
X_test['CPI_scaled'] = (X_test['CPI'] - X_test_cpi_mean) / X_test_cpi_std
```

### Random walk with drift

```python
 ## Make an array to hold the cv maes here
maes=np.zeros(5)
rmses= np.zeros(5)

## loop through the cv splits here
i = 0
for train_index, test_index in cv.split(X_train):
    ## Get the training and holdout sets
    X_tt = X_train.iloc[train_index]
    X_ho = X_train.iloc[test_index]
    y_tt = y_train.iloc[train_index]
    y_ho = y_train.iloc[test_index]
    
    ## Fit your model/Make your prediction on the holdout set here
    beta_hat = y_tt.diff()[1:].mean()

    tt_drift = y_tt.values[-1] + beta_hat*np.arange(1,len(y_ho) + 1)
    
    maes[i]=mean_absolute_error(y_ho, tt_drift)
                                
    rmses[i]=np.sqrt(mean_squared_error(y_ho, tt_drift))
    
    i = i + 1
```

### SARIMAX

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller

# Check stationarity of NASDAQ returns (your target)
def check_stationarity(series):
    result = adfuller(series.dropna())
    return result[1] < 0.05  # p-value threshold

# Example for your log returns
is_stationary = check_stationarity(y_train)
print(f"Target is stationary: {is_stationary}")
```

```python
# p, q, P, Q hyperparameter tuning
import pmdarima as pm

model = pm.auto_arima(
    y_train,
    exogenous=X_train[features],
    seasonal=True,
    m=5,                    # Weekly seasonality
    d=0,                    # Force no differencing
    D=0,                    # Force no seasonal differencing
    stepwise=True,          # Faster search
    trace=True,
    error_action='ignore',
    suppress_warnings=True,
    max_order=6,            # p+q+P+Q ≤ 6 to avoid overfitting
    information_criterion='aic'
)
print(model.order, model.seasonal_order)
```

```python
maes = np.zeros(5)
rmses= np.zeros(5)

order = (1, 0, 1)  # (p,d,q)
seasonal_order = (1, 0, 2, 5)  # (P,D,Q,s)

i = 0
for train_index, test_index in cv.split(X_train):
    X_tt = X_train.iloc[train_index]
    X_ho = X_train.iloc[test_index]
    y_tt = y_train.iloc[train_index]
    y_ho = y_train.iloc[test_index]
    
    ## Fit your model/Make your prediction on the holdout set
    model = SARIMAX(y_tt,
                   exog=X_tt[features],
                   order=order,
                   seasonal_order=seasonal_order,
                   trend='c')
    results = model.fit(disp=False)
    
    # Forecast (requires exog for test period)
    pred = results.get_forecast(steps=len(test_index), 
                             exog=X_ho[features])
    
    maes[i] = mean_absolute_error(y_ho, pred.predicted_mean)
    rmses[i]=np.sqrt(mean_squared_error(y_ho, pred.predicted_mean))                                   
    
    i += 1
```

### XGBoost

```python
# Define parameter grid (focused on key financial data parameters)
param_grid = {
    'max_depth': [3, 5],            # Tree depth (3 is conservative for finance)
    'learning_rate': [0.01, 0.05],   # Smaller rates for noisy returns
    'subsample': [0.8, 0.95],        # Prevents overfitting
    'colsample_bytree': [0.8, 1.0],  # Feature subsampling
    'n_estimators': [500, 1000]      # Number of trees
}
```

```python
# Initialize model with fixed params
model = XGBRegressor(
    objective='reg:squarederror',
    eval_metric='mae',
    early_stopping_rounds=50,
    random_state=42
)
```

```python
from sklearn.model_selection import GridSearchCV
# Grid search with time-series CV
grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    cv=cv,
    scoring='neg_mean_absolute_error',  # Use MAE for financial returns
    verbose=3,
    n_jobs=-1                          # Parallelize
)
```

### LSTM

```python
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)
        
    def forward(self, x):
        out, _ = self.lstm(x)
        out = self.fc(out[:, -1, :])  # Take last timestep
        return out
```

```python
# Convert data to PyTorch tensors
def create_sequences(X, y, seq_length):
    X_seq, y_seq = [], []
    for i in range(len(X) - seq_length):
        X_seq.append(X.iloc[i:i+seq_length].values)
        y_seq.append(y.iloc[i + seq_length])  
    return np.array(X_seq), pd.Series(y_seq)
```

```python
maes = np.zeros(5)
rmses= np.zeros(5)
seq_length= 20  # month
i = 0

for train_idx, val_idx in cv.split(X_train):
    # Split data
    X_tt, X_ho = X_train.iloc[train_idx], X_train.iloc[val_idx]
    y_tt, y_ho = y_train.iloc[train_idx], y_train.iloc[val_idx]
    
    # Create sequences
    X_seq_train, y_seq_train = create_sequences(X_tt[features], y_tt, seq_length)
    X_seq_val, y_seq_val = create_sequences(X_ho[features], y_ho, seq_length)
    
    # Convert to tensors
    X_seq_train = torch.tensor(X_seq_train, dtype=torch.float32)
    y_seq_train = torch.tensor(y_seq_train.values, dtype=torch.float32)
    X_seq_val = torch.tensor(X_seq_val, dtype=torch.float32)
    y_seq_val = torch.tensor(y_seq_val.values, dtype=torch.float32)
    
    # Reinitialize model and optimizer per fold
    lstm_model = LSTMModel(input_size=len(features), hidden_size=50, num_layers=2)
    optimizer = torch.optim.Adam(lstm_model.parameters(), lr=0.001)
    criterion = nn.MSELoss()
    
    # Train
    train_loader = DataLoader(TensorDataset(X_seq_train, y_seq_train), batch_size=32, shuffle=False)
    
    for epoch in range(50):
        lstm_model.train()
        for batch_x, batch_y in train_loader:
            optimizer.zero_grad()
            outputs = lstm_model(batch_x)
            loss = criterion(outputs, batch_y.unsqueeze(1))
            loss.backward()
            optimizer.step()
    
    # Validate
    lstm_model.eval()
    with torch.no_grad():
        lstm_pred = lstm_model(X_seq_val).squeeze().numpy()
        true_vals = y_seq_val.numpy()
        maes[i] = mean_absolute_error(true_vals, lstm_pred)
        rmses[i]= np.sqrt(mean_squared_error(true_vals,lstm_pred))
                                        
    i += 1
```
